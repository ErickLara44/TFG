# Adding spatio-temporal features and creating the final **IberFire** datacube:


This notebook will generate the final **IberFire** datacube, by adding the remaining features not added on ``2.1-Datacube-creation.ipynb``.


#### Note:

All the features merged here had previously some transformations applied in order to be homogeneized into the **IberFire**'s format.

```python
import xarray as xr

 
def datacube(chunks = None):
    """
    Opens the different variables and returns the datacube. Uses xr.open_mfdataset for that.
    """

    file_paths = []

    # Base datacube
    path = "....../Datacube variables peninsula"
    file_paths.append("......../Datacube/Base_datacube.nc")

    # Vegetation
    folder_path = path + "/Vegetation_variables"
    file_paths.append(folder_path + "/SoilWaterIndex_001_withoutNA.nc")
    file_paths.append(folder_path + "/SoilWaterIndex_005_withoutNA.nc")
    file_paths.append(folder_path + "/SoilWaterIndex_010_withoutNA.nc")
    file_paths.append(folder_path + "/SoilWaterIndex_020_withoutNA.nc")
    file_paths.append(folder_path + "/LST_withoutNA.nc")
    file_paths.append(folder_path + "/fapar_withoutNA.nc")
    file_paths.append(folder_path + "/lai_withoutNA.nc")
    file_paths.append(folder_path + "/ndvi_withoutNA.nc")

    # ERA5-Land
    folder_path = path + "/ERA5Land_variables"
    file_paths.append(folder_path + "/2mtemp_mean_celsius.nc")
    file_paths.append(folder_path + "/2mtemp_max_celsius.nc")
    file_paths.append(folder_path + "/2mtemp_min_celsius.nc")
    file_paths.append(folder_path + "/2mtemp_range_celsius.nc")
    
    file_paths.append(folder_path + "/surface_pressure_mean_hPa.nc")
    file_paths.append(folder_path + "/surface_pressure_max_hPa.nc")
    file_paths.append(folder_path + "/surface_pressure_min_hPa.nc")
    file_paths.append(folder_path + "/surface_pressure_range_hPa.nc")

    file_paths.append(folder_path + "/total_precipitation_mean_mm.nc")

    file_paths.append(folder_path + "/max_WS_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/mean_WS_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/mean_WD_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/wind_direction_at_max_speed_reprojected.nc")

    file_paths.append(folder_path + "/RH_max_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/RH_mean_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/RH_min_fromhourly_reprojected.nc")
    file_paths.append(folder_path + "/RH_range.nc")

    if chunks is not None:
        ds_combined =  xr.open_mfdataset(file_paths, combine='by_coords', chunks=chunks)
    else: 
        ds_combined =  xr.open_mfdataset(file_paths, combine='by_coords')

    ds_combined["LST"].attrs = {
    "long_name": "Land Surface Temperature",
    "description": "Temperature of the surface of the Earth.",
    "source": "Three different sources were used. From 2007/12/01 to 2010/06/19 the 'skin temperature' variable from ERA5Land was used: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview. From 2010/06/20 to 2021/01/18 the data was downloaded from the following Copernicus Land Monitoring Service variable: https://land.copernicus.eu/en/products/temperature-and-reflectance/hourly-land-surface-temperature-global-v1-0-5km. For the other dates, the source was the following variable from the CLMS: https://land.copernicus.eu/en/products/temperature-and-reflectance/hourly-land-surface-temperature-global-v2-0-5km.",
    "units": "Kelvin"
    }
    ds_combined["SWI_001"].attrs = {
        "long_name": "Soil Water Index at T=1",
        "description": "Provides daily updates on the moisture conditions in different soil layers.",
        "source": "Copernicus Land Monitoring Service: https://land.copernicus.eu/en/products/soil-moisture/daily-soil-water-index-global-v3-0-12-5km#download.",
        "units": "percentage"
    }
    ds_combined["SWI_005"].attrs = {
        "long_name": "Soil Water Index at T=5",
        "description": "Provides daily updates on the moisture conditions in different soil layers.",
        "source": "Copernicus Land Monitoring Service: https://land.copernicus.eu/en/products/soil-moisture/daily-soil-water-index-global-v3-0-12-5km#download.",
        "units": "percentage"
    }
    ds_combined["SWI_010"].attrs = {
        "long_name": "Soil Water Index at T=10",
        "description": "Provides daily updates on the moisture conditions in different soil layers.",
        "source": "Copernicus Land Monitoring Service: https://land.copernicus.eu/en/products/soil-moisture/daily-soil-water-index-global-v3-0-12-5km#download.",
        "units": "percentage"
    }
    ds_combined["SWI_020"].attrs = {
        "long_name": "Soil Water Index at T=20",
        "description": "Provides daily updates on the moisture conditions in different soil layers.",
        "source": "Copernicus Land Monitoring Service: https://land.copernicus.eu/en/products/soil-moisture/daily-soil-water-index-global-v3-0-12-5km#download.",
        "units": "percentage"
    }
    ds_combined["lai"].attrs = {
        "long_name": "Leaf Area Index", 
        "description": "Defined as half the total area of green elements of the canopy per unit horizontal ground area.",
        "source": "Two different datasets from the Copernicus Land Monitoring Services were used. The first one has 1km x 1km spatial resolution: https://land.copernicus.eu/en/products/vegetation/leaf-area-index-v2-0-1km. The second one has 300m x 300m spatial resolution: https://land.copernicus.eu/en/products/vegetation/leaf-area-index-300m-v1.0. The first source was used until 2020/04/30, and the second one was used up until 2024/12/31.",
        "units": "m^2/m^2"
    }
    ds_combined["fapar"].attrs = {
        "long_name": "Fraction of Absorbed Photosynthetically Active Radiation",
        "description": "Quantifies the fraction of the solar radiation absorbed by live plants for photosynthesis.",
        "source": "Two different datasets from the Copernicus Land Monitoring Services were used. The first one has 1km x 1km spatial resolution: https://land.copernicus.eu/en/products/vegetation/fraction-of-absorbed-photosynthetically-active-radiation-v2-0-1km. The second one has 300m x 300m spatial resolution: https://land.copernicus.eu/en/products/vegetation/fraction-of-absorbed-photosynthetically-active-radiation-v1-0-300m. The first source was used until 2020/04/30, and the second one was used up until 2024/12/31.",
        "units": "dimensionless"
    }
    ds_combined["ndvi"].attrs = {
        "long_name": "Normalized Difference Vegetation Index",
        "description": "NDVI is an indicator of the greenness of the biomes.",
        "source": "Two different datasets from the Copernicus Land Monitoring Services were used. The first one has 1km x 1km spatial resolution: https://land.copernicus.eu/en/products/vegetation/normalised-difference-vegetation-index-v3-0-1km#general_info. The second one has 300m x 300m spatial resolution: https://land.copernicus.eu/en/products/vegetation/normalised-difference-vegetation-index-v2-0-300m. The first source was used until 2020/06/21, and the second one was used up until 2024/12/31.",
        "units": "dimensionless"
    }
    ds_combined["max_rh"].attrs = {
    "long_name": "Maximum Relative Humidity",
    "description": "Relative Humidity is the percentage that indicates how much water vapor is in the air compared to the maximum amount the air can hold at a given temperature, before reaching saturation.",
    "source": "Derived from ERA5Land data using the 2m dewpoint temperature and 2m temperature.",
    "units": "percentage"
    }
    ds_combined["mean_rh"].attrs = {
        "long_name": "Mean Relative Humidity",
        "description": "Relative Humidity is the percentage that indicates how much water vapor is in the air compared to the maximum amount the air can hold at a given temperature, before reaching saturation.",
        "source": "Derived from ERA5Land data using the 2m dewpoint temperature and 2m temperature.",
        "units": "percentage"
    }
    ds_combined["min_rh"].attrs = {
        "long_name": "Minimum Relative Humidity",
        "description": "Relative Humidity is the percentage that indicates how much water vapor is in the air compared to the maximum amount the air can hold at a given temperature, before reaching saturation.",
        "source": "Derived from ERA5Land data using the 2m dewpoint temperature and 2m temperature.",
        "units": "percentage"
    }



    ds_combined["max_t2m"].attrs = {
        "long_name": "Maximum 2 Meter Temperature",
        "description": "Temperature of air at 2m above the surface of land, sea or in-land waters.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Celsius"
    }
    ds_combined["mean_t2m"].attrs = {
        "long_name": "Mean 2 Meter Temperature",
        "description": "Temperature of air at 2m above the surface of land, sea or in-land waters.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Celsius"
    }
    ds_combined["min_t2m"].attrs = {
        "long_name": "Minimum 2 Meter Temperature",
        "description": "Temperature of air at 2m above the surface of land, sea or in-land waters.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Celsius"
    }
    ds_combined["range_t2m"].attrs = {
        "long_name": "2m temperature range.",
        "description": "Range of 2m temperature.",
        "source": "Derived from ERA5Land data." ,
        "units": "Celsius"

    }

    ds_combined["max_surface_pressure"].attrs = {
        "long_name": "Maximum Surface Pressure",
        "description": "Pressure (force per unit area) of the atmosphere on the surface of land, sea and in-land water.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Hectopascals"
    }
    ds_combined["mean_surface_pressure"].attrs = {
        "long_name": "Mean Surface Pressure",
        "description": "Pressure (force per unit area) of the atmosphere on the surface of land, sea and in-land water.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Hectopascals"
    }
    ds_combined["min_surface_pressure"].attrs = {
        "long_name": "Minimum Surface Pressure",
        "description": "Pressure (force per unit area) of the atmosphere on the surface of land, sea and in-land water.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/derived-era5-land-daily-statistics?tab=overview.",
        "units": "Hectopascals"
    }
    ds_combined["range_surface_pressure"].attrs = {
        "long_name": "Surface pressure range.",
        "description": "Range of surface pressure.",
        "source": "Derived from ERA5Land data." ,
        "units": "Hectopascals"

    }

    ds_combined = ds_combined.rename({"total_precipitation": "mean_total_precipitation"})
    
    ds_combined["mean_total_precipitation"].attrs = {
        "long_name": "Mean Total Precipitations",
        "description": "Mean of the hourly values of the total precipitation variable from ERA5Land.",
        "source": "ERA5Land: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land?tab=overview",
        "units": "Millimeters"
    }

    ds_combined["max_wind_speed"].attrs = {
        "long_name": "Maximum Wind Speed",
        "description": "The highest wind speed derived from the hourly u-component and v-component of wind.",
        "source": "Derived from ERA5Land data: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land?tab=overview.",
        "units": "m/s"
    }
    ds_combined["mean_wind_speed"].attrs = {
        "long_name": "Mean Wind Speed",
        "description": "The average wind speed derived from the hourly u-component and v-component of wind.",
        "source": "Derived from ERA5Land data: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land?tab=overview.",
        "units": "m/s"
    }
    ds_combined["mean_wind_direction"].attrs = {
        "long_name": "Mean Wind Direction",
        "description": "The average wind direction derived from the average u-component and v-component of wind. The wind direction indicates the direction from where the wind comes.",
        "source": "Derived from ERA5Land data: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land?tab=overview.",
        "units": "Degrees"
    }
    ds_combined["wind_direction_at_max_speed"].attrs = {
        "long_name": "Wind Direction at Maximum Wind Speed",
        "description": "The wind direction where the maximum wind speed happened. Derived after calculating the hourly wind speed from the hourly u-component and v-component of wind. The wind direction indicates the direction from where the wind comes.",
        "source": "Derived from ERA5Land data: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land?tab=overview.",
        "units": "Degrees"
    }
    
    # Rename the features
    ds_combined = ds_combined.rename({f'popdens_{year}_mean': f'popdens_{year}' for year in range(2008, 2021)})
    ds_combined = ds_combined.rename(
        {"fapar": "FAPAR",
        "lai": "LAI",
        "ndvi": "NDVI"}
    )
    ds_combined = ds_combined.rename(
        {"mean_t2m": "t2m_mean",
         "max_t2m": "t2m_max",
         "min_t2m": "t2m_min",
         "range_t2m": "t2m_range",
         
         "mean_rh": "RH_mean",
         "max_rh": "RH_max",
         "min_rh": "RH_min",
         "range_rh": "RH_range",
         
         "max_surface_pressure": "surface_pressure_max",
         "mean_surface_pressure": "surface_pressure_mean",
         "min_surface_pressure": "surface_pressure_min",
         "range_surface_pressure": "surface_pressure_range",
         
         "max_wind_speed": "wind_speed_max",
         "mean_wind_speed": "wind_speed_mean",
         "mean_wind_direction": "wind_direction_mean",
         "wind_direction_at_max_speed": "wind_direction_at_max_speed",
         
         "mean_total_precipitation": "total_precipitation_mean",
         }
    )

    # Reorder the features
    ds_combined = ds_combined.drop_vars("crs")
    ordered_vars = []
    ordered_vars += ["x_index", "y_index", "is_spain"]
    ordered_vars += ["is_fire", "is_near_fire"]
    ordered_vars += ["x_coordinate", "y_coordinate", "is_sea", "is_waterbody", "AutonomousCommunities"]
    for year in [2006, 2012, 2018]:
        ordered_vars += [f"CLC_{year}_{i}" for i in range(1,45)]
        ordered_vars += [f"CLC_{year}_urban_fabric_proportion",
                         f"CLC_{year}_industrial_proportion",
                         f"CLC_{year}_mine_proportion",
                         f"CLC_{year}_artificial_vegetation_proportion",
                         f"CLC_{year}_arable_land_proportion",
                         f"CLC_{year}_permanent_crops_proportion",
                         f"CLC_{year}_heterogeneous_agriculture_proportion",
                         f"CLC_{year}_forest_proportion",
                         f"CLC_{year}_scrub_proportion",
                         f"CLC_{year}_open_space_proportion",
                         f"CLC_{year}_inland_wetlands_proportion",
                         f"CLC_{year}_maritime_wetlands_proportion",
                         f"CLC_{year}_inland_waters_proportion",
                         f"CLC_{year}_marine_waters_proportion"]
        ordered_vars += [f"CLC_{year}_artificial_proportion",
                        f"CLC_{year}_agricultural_proportion",
                        f"CLC_{year}_forest_and_semi_natural_proportion",
                        f"CLC_{year}_wetlands_proportion",
                        f"CLC_{year}_waterbody_proportion"]
    ordered_vars += [f"aspect_{i}" for i in range(1,9)] + ["aspect_NODATA"]
    ordered_vars += ["elevation_mean", "elevation_stdev", "slope_mean", "slope_stdev", "roughness_mean", "roughness_stdev"]
    ordered_vars += ["dist_to_roads_mean", "dist_to_roads_stdev", 
                     "dist_to_waterways_mean", "dist_to_waterways_stdev",
                     "dist_to_railways_mean", "dist_to_railways_stdev",
                     "is_natura2000"]
    ordered_vars += [f"popdens_{year}" for year in range(2008, 2021)]
    ordered_vars += ["is_holiday"]
    ordered_vars += ["t2m_mean", "t2m_max", "t2m_min", "t2m_range",
                     "RH_mean", "RH_max", "RH_min", "RH_range",
                     "surface_pressure_mean", "surface_pressure_max", "surface_pressure_min", "surface_pressure_range",
                     "total_precipitation_mean",
                     "wind_speed_mean", "wind_speed_max",
                     "wind_direction_mean", "wind_direction_at_max_speed"]
    ordered_vars += ["FAPAR", "LAI", "NDVI", "LST",
                     "SWI_001", "SWI_005", "SWI_010", "SWI_020"]
    ds_combined = ds_combined[ordered_vars]

    # Add attributes
    ds_combined.attrs = {
        "title": "IberFire",
        "description": "Datacube centered in Spain with 1km x 1km spatial resolution and daily temporal resolution.",
        "dimensions": "(y: 920, x: 1188, time: 6241)",
        "spatial_resolution": "1km x 1km",
        "temporal_resolution": "Daily",
        "start_date": "2007-12-01",
        "end_date": "2024-12-31",
        "crs": "EPSG:3035",
        "units_x": "meters",
        "units_y": "meters",
        "geospatial_x_min": ds_combined.x.min().values.item(),
        "geospatial_x_max": ds_combined.x.max().values.item(),
        "geospatial_y_min": ds_combined.y.min().values.item(),
        "geospatial_y_max": ds_combined.y.max().values.item(),
        "author": "Julen Ercibengoa Calvo",
        "author_contact": "julen.ercibengoa@gmail.com, julen.ercibengoa@tekniker.es",
        "creation_date": "2025-04-04",
    }
    
    return ds_combined

ds = datacube()
ds.to_netcdf("........./Datacube/IberFire.nc")

```
